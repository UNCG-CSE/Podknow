```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Reading csv data in as dataframe.

```python
data = pd.read_csv("..\\..\\data\\wc-cs.csv")
data.head()
```

Quick Stats

```python
X = data['WordCount'].iloc[0:1000].values.reshape(-1, 1)
y = data['ConfScore'].iloc[0:1000]

x_test = data['WordCount'].iloc[1000:2000].values.reshape(-1, 1)

print("X mean: " + str(X.mean()))
```

Plotting data as is

```python

plt.plot(X, y, 'o', alpha=0.10)
```

Random forest regression model

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X, y)
X_fit = np.linspace(0, 275, 250)[:, np.newaxis]
y_fit = model.predict(x_test)

plt.plot(x_test, y_fit, 'o', alpha=0.10);
plt.axis([0, 250, 0.75, 1.0])
```

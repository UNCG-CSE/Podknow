#using nltk to remove stop words  and form a excel sheet.

```python
from nltk.tokenize import word_tokenize
import string
from nltk.corpus import stopwords
```

```python
results = []
import pandas as pd
for folderName,subfolders,fileName in os.walk('C:\\Users\\Harini\\Final_Transcripts\\gcsst\\raw'):
       for file in fileName:
            if  file.endswith(".txt"):
                scrubbed = []
                #splitting the file name at '_'
                value0, value1, value2,value3,*extraWords = file.split('_')
                file = open(os.path.join(folderName,file),'rt')
                text = file.read() 
                tokens = word_tokenize(text)
                #converting to lower case and removing punctuation
                tokens = [w.lower() for w in tokens]
                table = str.maketrans('', '', string.punctuation)
                stripped = [w.translate(table) for w in tokens]
                #removin alpha numeric values
                words = [word for word in stripped if word.isalpha()]
                stop_words = set(stopwords.words('english'))
                #removing the stop words 
                words = [w for w in words if not w in stop_words]
                scrubbed.append(words)
                value4 = scrubbed
                #value5 needs to be changed based on the source
                value5 = 'gcsst'
                rows = (value0,value1,value2,value3,value4,value5)
                results.append(rows)
                    
                
               

labels = ['ID','Name','Date','topicname','scrubbedtext','ContentFrom']
PodKnow_Data = pd.DataFrame.from_records(results, columns= labels)

print(PodKnow_Data)
```
